```python
import numpy
import os
```

```python
distance_file = os.path.join('data','distance_data_headers.csv')
distances = numpy.genfromtxt(fname=distance_file, delimiter=',', dtype='unicode')
```

```python
headers = distances[0]
```

```python
data = distances[1:]
data = data.astype(numpy.float) # this is the recasting meaning we are changing the type of that data
```

```python
print(data)
```

```python
import matplotlib.pyplot as plt #All these modules and libraries are installed in Anaconda.
```

```python
plt.figure() # initialize the plot
plt.plot(data[:,0], data[:,1])
```

```python
#labeling the axes:
sample = headers[1]
plt.figure()
plt.xlabel('Simulation Frame')
plt.ylabel('Distance(angstrom)')
fig_1 = plt.plot(data[:,0], data[:,1], label=sample)
plt.legend() # putting that small box inside the plot because we want to see what's what.
plt.savefig(F'{sample}.png') #Saves the plot in png format in the directory of our work
```

```python
#Plotting two things at the same plot
plt.figure()
plt.xlabel('Simulation Frame')
plt.ylabel('Distance(angstrom)')
plt.plot(data[:,0], data[:,1], label=headers[1])
plt.plot(data[:,0], data[:,2], label=headers[2])
plt.legend()
```

```python
# The clever way to plot all the columns:
plt.figure()
for col in range(1, len(headers)):
    fig = plt.plot(data[:,0], data[:,col],label=headers[col])
plt.xlabel('Simulation Frame')
plt.ylabel('Distance(angstrom)')
plt.legend()
plt.savefig('all_samples_300dpi.png', dpi=300)
```

```python
#To Have the plot above, in different figures using same for loop:
#plt.figure()
color_list = ['b', 'g', 'm', 'k', 'c'] # To save each figure in a new color. Notice that it starts from 0 and does not consider the first color. So there should be one more color there.

for col in range(1, len(headers)):
    plt.figure() # if this line is out of for loop, then it adds the next plt to the same thing and saves new figure. So I initialized this in the loop.
    fig = plt.plot(data[:,0], data[:,col], color_list[col], label=headers[col])
    plt.xlabel('Simulation Frame')
    plt.ylabel('Distance(angstrom)')
    plt.legend(loc=1) # the comment in paranthesis is to fix the location of legend
    plt.savefig(F'{headers[col]}.png', dpi=300) # dpi defines the resolution
```
